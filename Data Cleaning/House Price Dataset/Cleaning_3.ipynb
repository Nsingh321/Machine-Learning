# Numerical Missing Values Imputation by Class

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# load data
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum().sum()
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.info()
```

```python
missing_values_percent = df.isnull().sum() / df.shape[0] * 100 
```

```python
missing_value_col_gre_20 = missing_values_percent[missing_values_percent > 20].keys()
missing_value_col_gre_20
```

```python
# as these columns has too much missing values so drop them
df2_drop_clm = df.drop(columns = missing_value_col_gre_20)
df2_drop_clm.shape
```

```python
# only numeric data columns
numeric_columns = df2_drop_clm.select_dtypes(include=['int64' , 'float64'])
numeric_columns.shape
```

```python
numeric_columns.isnull().sum()
```

```python
num_var_missing = [var for var in numeric_columns if numeric_columns[var].isnull().sum() > 0]
```

```python
df[num_var_missing][df[num_var_missing].isnull().any(axis='columns')]
```

```python
# filling 'LotFrontage' using the reference from 'LotConfig'(which is a categorical variable)

df['LotConfig'].unique()
```

```python
# filling  the mean value in 'LotFrontage' using the class 'Inside'(in LotConfig column)
df[df.loc[:,'LotConfig'] == "Inside"]['LotFrontage'].replace(
                                                            np.nan,
                                                            df[df.loc[:,'LotConfig'] == "Inside"]['LotFrontage'].mean()
                                                            )
```

```python
# doing this process for every class in 'LotConfig'
df_copy = df.copy() #making a copy of original dataset
for var_class in df['LotConfig'].unique():
    df_copy.update(df[df.loc[:,'LotConfig'] == var_class]['LotFrontage'].replace(
                                                            np.nan,
                                                            df[df.loc[:,'LotConfig'] == var_class]['LotFrontage'].mean()
                                                            ))
```

```python
df_copy.isnull().sum() #'LotFrontage' is cleaned(0 null values)
```

```python
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt'] #need to fill missing values in these columns
cat_vars = ['LotConfig','MasVnrType','GarageType']  #taking the reference from these columns ,filling the missing values of respecticive classes
 
for cat_var , num_var_miss in zip(cat_vars,num_vars_miss):
    
    #take the unique value in each column
    for var_class in df[cat_var].unique(): 
        
        #take every row of a particular class and compute the mean in num_vars_miss columns and replace with null values
        df_copy.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(
                                                            np.nan,
                                                            df[df.loc[:,cat_var] == var_class][num_var_miss].mean()
                                                           ))
```

```python
df_copy[num_var_miss].isnull().sum() #but the null values are still there
# because there we null values in categorical columns also
```

```python
df_copy[df_copy[['GarageType']].isnull().any(axis=1)]
```

```python
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt'] #need to fill missing values in these columns

# changing the 'MasVnrType','GarageType' to 'Exterior2nd','KitchenQual' as they were not cleaned
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']  #taking the reference from these columns ,filling the missing values of respecticive classes
 
for cat_var , num_var_miss in zip(cat_vars,num_vars_miss):
    
    #take the unique value in each column
    for var_class in df[cat_var].unique(): 
        
        #take every row of a particular class and compute the mean in num_vars_miss columns and replace with null values
        df_copy.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(
                                                            np.nan,
                                                            df[df.loc[:,cat_var] == var_class][num_var_miss].mean()
                                                           ))
```

```python
# missing values cleaned
df_copy[num_vars_miss].isnull().sum()
```

```python
# checking the data distribution to know the missing values that we imputed is fine oe not

plt.figure(figsize = (10,10))
sns.set()

for i , var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    
#     original dataset without imputing the values in column
    sns.distplot(df[var],bins=20,kde_kws={'linewidth':5} , label='Original')
    
#     after imputing mean values in the column
    sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':2} , label='Mean')
    plt.legend()
```

# Filling missing values with median

```python
df_copy_median = df.copy()

num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt'] #need to fill missing values in these columns

# changing the 'MasVnrType','GarageType' to 'Exterior2nd','KitchenQual' as they were not cleaned
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']  #taking the reference from these columns ,filling the missing values of respecticive classes
 
for cat_var , num_var_miss in zip(cat_vars,num_vars_miss):
    
    #take the unique value in each column
    for var_class in df[cat_var].unique(): 
        
        #take every row of a particular class and compute the mean in num_vars_miss columns and replace with null values
        df_copy_median.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(
                                                            np.nan,
                                                            df[df.loc[:,cat_var] == var_class][num_var_miss].median()
                                                           ))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize = (10,10))
sns.set()

for i , var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    
#     original dataset without imputing the values in column
    sns.distplot(df[var],bins=20,kde_kws={'linewidth':7} , label='Original')
    
#     after imputing mean values in the column
    sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':5} , label='Mean')
    
#     after imputing median values in the column
    sns.distplot(df_copy_median[var],bins=20,kde_kws={'linewidth':2} , label='Median')
    plt.legend()
```

```python
# to check for outliers we use boxplot
for i , var in enumerate(num_vars_miss):
    plt.figure(figsize=(10,10))
    
    # box plot of original
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    
    # box plot of dataset with mean values computed
    plt.subplot(3,1,2)
    sns.boxplot(df_copy[var])
    
    # box plot of dataset with median values computed
    plt.subplot(3,1,3)
    sns.boxplot(df_copy_median[var])
```
