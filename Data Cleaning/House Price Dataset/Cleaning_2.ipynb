# Imputing missing values with mean and median

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('train.csv')
df
```

```python
df.shape
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows'   ,None)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
missing_value_percent = df.isnull().sum()/ df.shape[0] *100
missing_value_percent
```

```python
# remove column which has more than 20% data missing
missing_value_col_gre_20 = missing_value_percent[missing_value_percent > 20].keys()
missing_value_col_gre_20
```

```python
df2_drop_clm = df.drop(columns = missing_value_col_gre_20)
df2_drop_clm.shape
```

```python
# only numeric columns
numeric_columns = df2_drop_clm.select_dtypes(include=['int64' , 'float64'])
numeric_columns.head()
```

```python
# plotting heatmap for visulizing the missing values 
plt.figure(figsize = (16,9))
sns.heatmap(data = numeric_columns.isnull())
```

```python
# to know the rows with missing values
numeric_columns[numeric_columns.isnull() .any(axis='columns')]
```

```python
numeric_columns.isnull().sum()
```

```python
# get the columns with missing values
missing_num_var = [var for var in numeric_columns.columns if numeric_columns[var].isnull().sum()>0]
missing_num_var
```

```python
# check data distribution to know when to use mean or median techniques to fill values
plt.figure(figsize=(10,10))

for i , var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(numeric_columns[var],bins=20,kde_kws = {'linewidth':3 , 'color':'#DC143C'})
    
# first plot is normal distributed so we can use mean,median or mode to computer the missing values
# second and third plot is skewed so we use median to compute the missing values
```

```python
# fill with the mean values where there is null values
df4_num_mean = numeric_columns.fillna(numeric_columns.mean())
df4_num_mean.isnull().sum()
```

```python
# data distribution b/w the original df and df4_num_mean
plt.figure(figsize = (10,10))

for i , var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var],bins=20,kde_kws={'linewidth':5} , label='Original')
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':2} , label='Final')
    plt.legend()
```

```python
# filling median values in numeric_columns dataframe
df5_median = numeric_columns.fillna(numeric_columns.median())
df5_median.isnull().sum()
```

```python
plt.figure(figsize = (10,10))
sns.set()
for i , var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var],bins=20,hist=False,kde_kws={'linewidth':8} , label='Original')
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':5} , label='Mean')
    sns.distplot(df5_median[var],bins=20,hist=False,kde_kws={'linewidth':3} , label='Median')
    plt.legend()
```

```python
# to check for outliers we use boxplot
for i , var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    
    # box plot of original
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    
    # box plot of dataset with mean values computed
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    
    # box plot of dataset with median values computed
    plt.subplot(3,1,3)
    sns.boxplot(df5_median[var])
```

```python
df6 = pd.concat([numeric_columns[missing_num_var], df4_num_mean[missing_num_var] , df5_median[missing_num_var]] , axis='columns')
```
