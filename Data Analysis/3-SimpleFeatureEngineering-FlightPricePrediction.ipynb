```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_excel('/kaggle/input/flight-price-prediction-dataset/Data_Train.xlsx')
df_test = pd.read_excel('/kaggle/input/flight-price-prediction-dataset/Test_set.xlsx')
df_train.head()
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
final_df = df_train.append(df_test)
print(final_df.shape)
```

```python
final_df.info()
```

```python
# Feature Engineering Process
# Extracting Day,Month,Year
final_df['Day'] = final_df['Date_of_Journey'].map(lambda x: x.split('/')[0]).astype(int)
final_df['Month'] = final_df['Date_of_Journey'].map(lambda x: x.split('/')[1]).astype(int)
final_df['Year'] = final_df['Date_of_Journey'].map(lambda x: x.split('/')[2]).astype(int)
```

```python
# Extracting Arrival Hour and Arrival Minute
final_df['Arrival_hour'] = final_df['Arrival_Time'].map(lambda x: x.split(' ')[0].split(':')[0]).astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Time'].map(lambda x: x.split(' ')[0].split(':')[1]).astype(int)
```

```python
# Extracting Dept Hour and Dept Minute
final_df['Dept_hour'] = final_df['Dep_Time'].map(lambda x: x.split(':')[0]).astype(int)
final_df['Dept_Minute'] = final_df['Dep_Time'].map(lambda x: x.split(':')[1]).astype(int)
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df['Total_Stops'].unique()
```

```python
# Handling categorical data in 'Total_Stops'
final_df['Total_Stops'] = final_df['Total_Stops'].map({
                            'non-stop':0,
                            '2 stops': 2,
                            '1 stop':1,
                            '3 stops':3,
                            '4 stops':4,
                            'nan': 2
                            })
```

```python
final_df.head()
```

```python
# Extracting Duration_hour 
final_df['Duration_hour'] = final_df['Duration'].map(lambda x : x.split(' ')[0].split('h')[0])
```

```python
print(final_df[final_df['Duration_hour']=='5m'])
```

```python
# Removing these 2 rows
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_min'] = final_df['Duration_hour'].map(lambda x: x*60)
```

```python
final_df.drop(['Duration','Dep_Time','Date_of_Journey','Arrival_Time','Route'],axis=1,inplace=True)
```

```python
final_df['Airline'].unique()
```

```python
# Applying Label Encoding on 'Airline','Source','Destination','Additional_Info'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
final_df['Airline'] = le.fit_transform(final_df['Airline'])
final_df['Source'] = le.fit_transform(final_df['Source'])
final_df['Destination'] = le.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])
```
